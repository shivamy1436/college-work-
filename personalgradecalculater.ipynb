```python
import numpy as np

```

```python
test_scores = np.array((5,7,8,2,1))
print(test_scores)
```

```python
# mean , meadian,std
mean = 0
for x in test_scores:
    # print(x)
    mean = mean + x

print(mean/len(test_scores)); print(test_scores.mean())
```

```python
median = len(test_scores)/2
print(int(median))
print(test_scores[int(median)])
```

```python
my_mean = test_scores.mean()
actual_value = 0
differ = (test_scores - my_mean) ** 2 
print(differ)
squared_value  = differ.mean()
actual_value = np.sqrt(squared_value)
print(actual_value); print(np.std(test_scores))
    
```
